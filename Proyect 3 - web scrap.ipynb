```python
from bs4 import BeautifulSoup
from splinter import Browser
import pandas as pd
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://spotifycharts.com/regional'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
names = soup.find_all('li', class_='responsive-select')
```

```python
sele = soup.select('li')
```

```python
data = []
country = []

for x in sele:
    data.append(x.attrs.get("data-value", None))
    country.append(x.text)
```

```python
data2 = data[1:66]
```

```python
country2 = country[1:66]
```

```python
for x in data2:
    url_dinamic = f'https://spotifycharts.com/regional/{x}/weekly/latest'
    browser.visit(url_dinamic)
```

```python
lista = list(zip(data2, country2))
df = pd.DataFrame(lista, columns = ['Id', 'Country'])
```

```python
us = pd.read_csv('../DataSource/csv/regional-us-weekly-latest.csv', skiprows=1)
us['Country'] = 'United States'
```

```python
for index, row in df[1:].iterrows():
    clave = row[0]
    pai = row[1]
    try:
        csv = pd.read_csv(f'../DataSource/csv/regional-{clave}-weekly-latest.csv', skiprows=1)
        csv['Country'] = str(pai)
        us = pd.concat([us, csv])
    except:
        print(f'No se encontro el archivo {clave}, con el país {pai}')
```

```python
id = []

for index, row in us.iterrows():
    separado = row[4].split("/")
    id.append(separado[4])
    
us['Track Id'] = id
us
```

```python
us.to_csv('Listas.csv', index=False)
```
